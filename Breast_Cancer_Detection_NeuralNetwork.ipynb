## **Importing the Dependencies**

## **Neural Network Implementation**

Splitting the data into Target and Feature variables.

```python
X = df.drop(columns='diagnosis', axis=1)
y = df['diagnosis']
```

Splitting the data into train and test

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 88)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Standardizing the data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = Sequential()
model.add(Flatten(input_shape=(30,)))  # Flatten layer to convert our data to 1D array
model.add(Dense(15, activation='relu'))  # Hidden Layer
model.add(Dense(2, activation='sigmoid')) # Output layer, where 2 neuron specifies we want 2 outputs
```

```python
# compiling the neural network

model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics= ['accuracy'])
# Here The target labels are represented as integers. For example, if we have 3 classes, the labels could be 0, 1, or 2.
# If we had one hot encoding, we would have used categorical_cross_entropy
```

```python
history = model.fit(X_train, y_train, epochs = 10, validation_split = 0.2)
```

Visualizing Accuracy and Loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('Model Accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'])
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('Model Loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'])
plt.show()
```

Accuracy of the model on the Test data

```python
loss, accuracy = model.evaluate(X_test, y_test)

print(f"Loss on test data: {loss:.4f}")
print(f"Accuracy on test data: {accuracy:.4f}")
```

```python
y_pred_model = model.predict(X_test)
```
